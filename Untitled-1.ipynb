```python
# Data initialization
users = {}
current_user = None
cart = []

menu = {
    "Veg": {
        "Snacks": [("Samosa", 10), ("Veg Puff", 20), ("Paneer Pakora", 25), ("Aloo Tikki", 15)],
        "Starters": [("Paneer Tikka", 100), ("Veg Manchurian", 120), ("Hara Bhara Kabab", 90), ("Chilli Paneer", 110)],
        "Main Course": [("Veg Biryani", 150), ("Paneer Butter Masala", 200), ("Aloo Gobi", 130), ("Chole Bhature", 140)],
    },
    "Non-Veg": {
        "Snacks": [("Chicken Roll", 50), ("Chicken Puff", 30), ("Mutton Samosa", 40), ("Fish Cutlet", 35)],
        "Starters": [("Chicken Tikka", 150), ("Fish Fingers", 180), ("Mutton Seekh Kabab", 200), ("Prawn Tempura", 220)],
        "Main Course": [("Chicken Biryani", 200), ("Butter Chicken", 250), ("Mutton Rogan Josh", 280), ("Fish Curry", 240)],
    },
}

def register():
    while True:
        username = input("Enter username: ")
        if username in users:
            print("Username already exists. Please try again.")
        else:
            break
    password = input("Enter password: ")
    users[username] = password
    print("Registration successful.")

def login():
    global current_user
    while True:
        username = input("Enter username: ")
        password = input("Enter password: ")
        if username in users and users[username] == password:
            current_user = username
            print("Login successful.")
            main_menu()
            break
        else:
            print("Invalid username or password. Please try again.")

def main_menu():
    options = [("Dine-In", food_category_selection), ("Delivery", food_category_selection), ("Logout", logout)]
    while True:
        print("\nMain Menu")
        for i, (option, _) in enumerate(options, 1):
            print(f"{i}. {option}")
        choice = input("Enter your choice: ")

        if choice.isdigit() and 1 <= int(choice) <= len(options):
            _, action = options[int(choice) - 1]
            action()
            if current_user is None:  # If the user logged out
                break
        else:
            print("Invalid choice. Please try again.")

def food_category_selection():
    categories = [("Veg", "Veg"), ("Non-Veg", "Non-Veg"), ("Back to Main Menu", None)]
    while True:
        print("\nFood Category")
        for i, (name, _) in enumerate(categories, 1):
            print(f"{i}. {name}")
        category_choice = input("Enter your choice: ")

        if category_choice.isdigit() and 1 <= int(category_choice) <= len(categories):
            category_name, category_key = categories[int(category_choice) - 1]
            if category_key:
                food_type_selection(category_key)
            else:
                break
        else:
            print("Invalid choice. Please try again.")

def food_type_selection(category):
    types = [("Snacks", "1"), ("Starters", "2"), ("Main Course", "3"), ("Back to Food Category", None)]
    while True:
        print(f"\nFood Type - {category}")
        for i, (name, _) in enumerate(types, 1):
            print(f"{i}. {name}")
        type_choice = input("Enter your choice: ")

        if type_choice.isdigit() and 1 <= int(type_choice) <= len(types):
            type_name, type_key = types[int(type_choice) - 1]
            if type_key:
                display_food_items(category, type_key)
            else:
                break
        else:
            print("Invalid choice. Please try again.")

def display_food_items(category, type_choice):
    food_type = {'1': 'Snacks', '2': 'Starters', '3': 'Main Course'}[type_choice]
    food_items = menu[category][food_type]

    if food_items:
        print(f"\nAvailable {food_type} - {category}")
        for idx, item in enumerate(food_items):
            print(f"ID: {idx + 1}, Name: {item[0]}, Price: ${item[1]}")

        while True:
            choice = input("Enter food ID to add to cart, or 'b' to go back: ")
            if choice == 'b':
                break
            elif choice.isdigit():
                food_id = int(choice) - 1
                if 0 <= food_id < len(food_items):
                    add_to_cart(food_items[food_id])
                else:
                    print("Invalid food ID. Please try again.")
            else:
                print("Invalid input. Please try again.")
    else:
        print(f"No {food_type} items available in {category} category.")

def add_to_cart(food_item):
    cart.append(food_item)
    print(f"Added {food_item[0]} to cart.")

def checkout():
    if not cart:
        print("Your cart is empty.")
        return

    print("\nYour Cart")
    total = 0
    for item in cart:
        print(f"Name: {item[0]}, Price: ${item[1]}")
        total += item[1]

    print(f"Total: ${total}")
    choice = input("Proceed to checkout? (yes/no): ")
    if choice.lower() == 'yes':
        print("Order placed successfully!")
        cart.clear()
    else:
        print("Checkout cancelled.")

def logout():
    global current_user
    current_user = None
    print("You have been logged out.")

def main():
    while True:
        print("\nWelcome to Zomato CLI")
        options = [("Login", login), ("Register", register), ("Exit", exit)]
        for i, (option, _) in enumerate(options, 1):
            print(f"{i}. {option}")
        choice = input("Enter your choice: ")

        if choice.isdigit() and 1 <= int(choice) <= len(options):
            _, action = options[int(choice) - 1]
            action()
        else:
            print("Invalid choice. Please try again.")

if __name__ == '__main__':
    main()
```
